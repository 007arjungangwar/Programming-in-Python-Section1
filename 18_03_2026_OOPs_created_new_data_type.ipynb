````python
class Atm:
    # if you wants to Declare The variable inside the class then you should have to declare inside the init function
    def __init__(self):  # Name of the constructor is fix
        self.pin=""
        self.balance=0
        self.menu()

    def menu(self):
        user_input=input("""
                        Hello Welcome ,How would you like to proceed?
                        Enter 1 to creare pin
                        Enter 2 to deposite
                        Enter 3 to Withdraw
                        Enter 4 to check balance
                        Enter 5 to exit.""")

        if user_input=="1":
            self.create_pin()
            # print("Pin created Successfully")
            self.menu()
        elif user_input=="2":
            self.deposite()
            # print("deposite Successful")
            self.menu()
        elif user_input=="3":
            self.withdraw()
            # print("Withdraw Successful")
            self.menu()
        elif user_input=="4":
            self.check_balance()
            # print(f"your balance is: {self.balance}")
            self.menu()
        elif user_input=="5":
            print("Thanks for Visiting")
        else:
          print(" Operation not allowed")

    def create_pin(self):
        user_input=input("Enter the pin for creating.")
        print("Pin Successfully created.")
        self.pin=user_input


    def deposite(self):
        user_input=input("Enter the pin for deposite.")
        if self.pin==user_input:
            temp=int(input("Enter amount for deposite."))
            self.balance+=temp
            print("deposite successful")
            print(f"Your updated balance is {self.balance}")
        else:
            print("Not Allowed")

    def withdraw(self):
        user_input=input("Enter the pin for withdraw.")
        print(self.pin)
        if user_input==self.pin:
            withdraw=int(input("Enter amount for withdraw."))
            if withdraw<=self.balance:
                self.balance-=withdraw
                print("Withdraw Successful")
                print(f"Your updated balance is {self.balance}")
            else:
                print("Insufficient fund")
        else:
            print("Not Allowed")


    def check_balance(self):
        temp=input("Enter pin for checking balance")
        if self.pin==temp:
            print(f"Your current balance is {self.balance}")
        else:
            print("Pin is not correct")
````

````python
# 4 feature of oops
"""
1. Abstraction - hiding unnecessary detail from user by method or Class
2. Encapsulation - Restrict access of certain attribute or methods to protect data and enforce controlled access (Data Protection) private
3. Inheritance - Reuse the property or method from the parent class
4. Polymorphism -  Allow method in different classes to have same name but different behaviour According to object Referred to class.
"""
````

## **Abstraction - hiding information by function(method or class)**

````python
# Abstraction - hiding information by function(method or class)

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.percentage=percentage

    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {round((self.percentage+5)/1.1,2)}%"

    # hiding info how pecentage is calculating

s2=Student("Roshni",12,92)
s2.student_info()
````

## **Encapsulation: - restrict access of some attribute or method**

````python
# Encapsulation: - restrict access of some attribute or method

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.__percentage=percentage

    def get_percentage(self):
        return self.__percentage

    def set_percentage(self,temp): # Instead of temp you can pass any of the name
        if 0<=temp<=100:
            self.__percentage=temp
            return self.__percentage
        else:
            return ("Not Allowed")


    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {self.get_percentage()}%"
````

## **Inheritance - inherit property of class by super**

````python
# Inheritance - inherit property of class by super

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.percentage=percentage

    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {round((self.percentage+5)/1.1,2)}%"


class GraduateStudent(Student):
    def __init__(self, name, grade, percentage, subject, marks, section, stream):
        super().__init__(name,grade,percentage)
        self.marks=marks
        self.section=section
        self.stream=stream
        self.subject=subject
        super().student_info()
````

````python
s1=GraduateStudent("Nitish",12,55,["Maths", "English", "Computer"],[95,92,95] ,"section2","PCM")
s2=GraduateStudent("Advika",11,98,["Maths", "English", "Bio"],[90,97,82] ,"section2","PCB")
````

````python
print(s1.subject)
print(s1.marks)
print(s1.grade)
print(s1.student_info())
````

````python
print(s2.subject)
print(s2.marks)
print(s2.grade)
print(s2.student_info())
````

## **Polymorphism : same method name but working is different**

````python
# Polymorphism : same method name but working is different

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.percentage=percentage

    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {round((self.percentage+5)/1.1,2)}%"


class GraduateStudent(Student):
    def __init__(self, name, grade, percentage, subject, marks, section, stream):
        super().__init__(name,grade,percentage)
        self.marks=marks
        self.section=section
        self.stream=stream
        self.subject=subject
        super().student_info()

    def student_info(self):
        return f"{self.name} was in Section {self.section} had subjects {self.subject[0], self.subject[1], self.subject[2]} and stream was {self.stream} pass out from the ABC Internetional with percentage {self.percentage}"
````

````python
class Fraction:
  def __init__(self,n,d):
    self.num=n
    self.den=d

  def __str__(self):
    return f"{self.num}/{self.den}"

  def __add__(self,other):
    temp_num=self.num*other.den+other.num*self.den
    temp_den=self.den*other.den
    return f"{temp_num}/{temp_den}"
````

````python
class Fraction:
  def __init__(self,n,d):
    self.num=n
    self.den=d

  def __str__(self):
    return f"{self.num}/{self.den}"

  def __add__(self,other):
    temp_num=self.num*other.den+other.num*self.den
    temp_den=self.den*other.den
    return f"{temp_num}/{temp_den}"

  def __sub__(self,other):
    temp_num=self.num*other.den-other.num*self.den
    temp_den=self.den*other.den
    return f"{temp_num}/{temp_den}"

  def __mul__(self,other):
    temp_num=self.num*other.num
    temp_den=self.den*other.den
    return f"{temp_num}/{temp_den}"

  def __truediv__(self,other):
    temp_num=self.num*other.den
    temp_den=self.den*other.num
    return f"{temp_num}/{temp_den}"

  def __pow__(self, power):
      temp_num = self.num ** power
      temp_den = self.den ** power
      return Fraction(temp_num, temp_den)
````

````python
x=Fraction(2,3)
print(x)

y=Fraction(3,4)
print(y)
````

````python
class Point:

  def __init__(self,x,y):
    self.x_cod = x
    self.y_cod = y

  def __str__(self):
    return '<{},{}>'.format(self.x_cod,self.y_cod)

  def euclidean_distance(self,other):
    return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5

  def distance_from_origin(self):
    return (self.x_cod**2 + self.y_cod**2)**0.5
    # return self.euclidean_distance(Point(0,0))
````
