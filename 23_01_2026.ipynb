# Operators in Python

- **Arithmetic**: `+`, `-`, `*`, `/`, `%`, `**`, `//`
- **Comparison**: `==`, `!=`, `>`, `<`, `>=`, `<=`
- **Logical**: `and`, `or`, `not`
- **Bitwise**: `&`, `|`, `^`, `~`, `<<`, `>>`
- **Assignment**: `=`, `+=`, `-=`, `*=`, `/=`, `%=`, `**=`, `//=`
- **Identity**: `is`, `is not`
- **Membership**: `in`, `not in`

## **Take the two number input from the user and return the their sum**




```python
# first we will take input from the user
num1= input(" Enter the first number : ")
num2= input(" Enter the second number : ")
# we will do their sum
total= num1+num2
# return their value
print("the sum of the given number : ", total)
```

```python
# first we will take input from the user
num1= int(input(" Enter the first number : "))
num2= int(input(" Enter the second number : "))
# we will do their sum
total= num1+num2
# return their value
print("the sum of the given two numbers : ", total)
```
