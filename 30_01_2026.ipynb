
# üîπ**Difference between == and is in Python**
```
== ‚Üí Equality Operator

üëâ Checks values are same or not
Compares data/value
Does NOT care about memory location

is ‚Üí Identity Operator

üëâ Checks memory location (object identity)
Compares object reference
Checks whether both variables point to the same object
```



````python
a = 10
b = 10
print(a is b)   # True

x = [1, 2, 3]
y = [1, 2, 3]
print(x is y)   # False
````

````python
data = None # Represent nothing available inside

if data is None:
    print("No data available")
````

# **Nested Conditional Statement**



```
4Ô∏è‚É£ Nested Conditional Statement
üîπ Definition

When an if statement is written inside another if or else block.

üîπ Syntax
if condition1:
    if condition2:
        statement(s)
    else:
        statement(s)
else:
    statement(s)

üîπ When to use

When a condition depends on another condition
For hierarchical decision-making
```



````python
age=int(input("Enter your age : "))

if age<18:
  print("Sorry you are Minnor can`t drive")
else:
  has_licence=input("You have Licence yes or No Please enter: ").upper()
  if age>18 and has_licence=="YES":
    print("You can drive")
  elif age>18 and has_licence=="NO":
    print("Sorry but driving licence is compulsory")
  else:
    print("Somthing is wrong, Please check your input once again")
````

````python
age = 20
id_card = True

if age >= 18:
    if id_card:
        print("Entry allowed")
    else:
        print("ID card required")
else:
    print("Underage")
````



```
Remember üß†

Indentation is very important
Avoid deep nesting (use elif when possible)
```

## **while Loop**
```
üîπ Definition : A while loop executes a block of code as long as a condition is True.

üîπ Syntax
while condition:
    statement(s)

üîπ When to use

When the number of iterations is NOT known
Loop depends on a condition

```



````python
i=0
while i <10:
  print(i)
  i=i+1
````



```
üîπ Remember üß†

Always update the condition variable
Otherwise ‚Üí infinite loop
```



````python
num=0
while num <10:
  print(num, end=" ")
  num=num+1
````
