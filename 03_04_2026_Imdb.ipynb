**Top 10 Most Frequent Genres**

```python
# Explode genres and count
top_genres = df['Genre'].str.split(', ').explode().value_counts().head(10)

plt.figure(figsize=(10,3))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')
plt.title('Top 10 Genres in IMDB Top 1000')
plt.xlabel('Number of Movies')
plt.show()
```

**IMDB Rating vs. Metascore (Critics vs. Fans)**

**This scatter plot compares the IMDB Rating (user-driven) against the Metascore (critic-given)**

```python
plt.figure(figsize=(10,3))
sns.scatterplot(data=df, x='IMDB_Rating', y='Meta_score', alpha=0.6)
plt.title('User Rating (IMDB) vs. Critic Rating (Metascore)')
plt.show()
```

**Distribution of Movie Runtime**

**A histogram showing how long the top-rated movies usually are.**

```python
plt.figure(figsize=(10,3))
sns.histplot(df['Runtime'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Movie Runtimes')
plt.xlabel('Runtime (minutes)')
plt.show()
```

**Average rating by genre**

```python
# Average rating by genre
df.groupby('Genre')['IMDB_Rating'].mean()
```

```python
df.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)
```

```python
# Average rating by genre
genre_rating = df.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 4))
genre_rating.plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Average IMDB Rating')
plt.title('Top 10 Genres by Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Rating trend over years**

```python
# Rating trend over years
df.groupby('Released_Year')['IMDB_Rating'].mean()
```

```python
# Rating trend over years
yearly_rating = df.groupby('Released_Year')['IMDB_Rating'].mean()

plt.figure(figsize=(25, 4))
plt.plot(yearly_rating.index, yearly_rating.values, marker='s', color='red')
plt.xlabel('Year')
plt.ylabel('Average IMDB Rating')
plt.title('Average IMDB Rating Trend Over Years')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Movies per year**

```python
df['Released_Year'].value_counts().sort_index()
```

```python
# Movies per year
year_counts = df['Released_Year'].value_counts().sort_index().head(50)

plt.figure(figsize=(12, 4))
plt.plot(year_counts.index, year_counts.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies in Top 1000 by Year (Last 50 Years)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
