- **Bitwise**: `&`, `|`, `^`, `~`, `<<`, `>>`




```
Bitwise Operators:

â€¢ | Bitwise OR
â€¢ ^ Bitwise XOR

```



````python
a=bin(2)
b=bin(3)
print("binary value : ",a)
print("binary value : ",b)
````

````python
x = 2 # Binary : 010
y = 3 # Binary : 011

print ( " x & y ( AND ) : " , x & y ) # 2 (010 & 011 = 010)
print ( " x | y ( OR ) : " , x | y ) # 3 (010 | 011 = 011)
print ( " x ^ y ( XOR ) : " , x ^ y ) # 1 (010 ^ 011 = 001)
print ( " ~ x ( NOT ) : " , ~ x ) # -3 ( Bitwise NOT )
print ( " x >> 2 ( Right Shift ) : " , x >> 2) # 0 (010 000)
print ( " y << 3 ( Left Shift ) : " , y << 3) # 24 (011 011000)
````

- **Identity**: `is`, `is not`




```
Identity Operators:
â€¢ is - Returns True if both variables point to the same object
â€¢ is not - Returns True if both variables point to different objects
```



````python
# Identity operators check if two variables refer to the same object
# For immutable types ( integers , strings, tuple )
a = 3
b = 3
print ("id ( a ) : ", id ( a ))
print ("id ( b ) : ", id ( b ))
print ("a is b (int) : " , a is b ) # True ( same object )

a = "Arjun"
b = "Arjun"
print ("id ( a ) : ", id ( a ))
print ("id ( b ) : ", id ( b ))
print ("a is b (string) : " , a is b ) # True ( string interning )
````

````python
# For mutable types ( lists , dictionaries )
a = [1 , 2 , 3]
b = [1 , 2 , 3]
print ("id ( a ) : ", id ( a ))
print ("id ( b ) : ", id ( b ))
print ("a is b ( list ) : " , a is b ) # False ( different objects )
print ("a == b ( list ) : " , a == b ) # True ( same content )
````

````python
# Complex strings may not be interned
a = " Hello - World "
b = " Hello - World "
print ("id ( a ) : ", id ( a ))
print ("id ( b ) : ", id ( b ))
print ("a is b ( complex string ) : " , a is b ) # May be True or False
print ("a is not b : " , a is not b ) # Opposite of â€™ is â€™
````


### **if-elif-else Statement**



```
3ï¸âƒ£ if-elif-else Statement
ðŸ”¹ Definition
Used to check multiple conditions in sequence.

ðŸ”¹ Syntax
if condition1:
    statement(s)
elif condition2:
    statement(s)
elif condition3:
    statement(s)
else:
    statement(s)

ðŸ”¹ When to use

When more than two conditions are required
Conditions are mutually exclusive
```



````python
age=int(input("Enter your Age : "))
#age=45
if age >20:
    print("You are Adult")
elif 12<=age<=19:
    print("Teenager")
else:
    print("You are a child")
````

````python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
````

```
4ï¸âƒ£ Nested Conditional Statement
ðŸ”¹ Definition

When an if statement is written inside another if or else block.

ðŸ”¹ Syntax
if condition1:
    if condition2:
        statement(s)
    else:
        statement(s)
else:
    statement(s)

ðŸ”¹ When to use

When a condition depends on another condition
For hierarchical decision-making
```



````python
age = 20
id_card = True

if age >= 18:
    if id_card:
        print("Entry allowed")
    else:
        print("ID card required")
else:
    print("Underage")
````



```
Remember ðŸ§ 

Indentation is very important
Avoid deep nesting (use elif when possible)
```
