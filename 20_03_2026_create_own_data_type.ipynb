# **Create new data type with the name of the Fraction**

```python
class Fraction:
    def __init__(self,n,d):
        self.num=n
        self.den=d

    # def __str__(self):
    #     return "{}/{}".format(self.num,self.den)

    def __str__(self):
        return f"{self.num}/{self.den}"

    def __add__(self,other):
        temp_num=self.num*other.den+other.num*self.den
        tem_den=self.den*other.den
        return "{}/{}".format(temp_num,tem_den)

    def __sub__(self,other):
        temp_num=self.num*other.den-other.num*self.den
        tem_den=self.den*other.den
        return "{}/{}".format(temp_num,tem_den)
```

### **Add some more functionality in the above fraction data type**

```python
class Fraction:
  def __init__(self,n,d):
    self.num=n
    self.den=d

  def __str__(self):
    return f"{self.num}/{self.den}"

  def __add__(self,other):
    temp_num=self.num*other.den+other.num*self.den
    temp_den=self.den*other.den
    return f"{temp_num}/{temp_den}"

  def __sub__(self,other):
    temp_num=self.num*other.den-other.num*self.den
    temp_den=self.den*other.den
    return f"{temp_num}/{temp_den}"

  def __mul__(self,other):
    temp_num=self.num*other.num
    temp_den=self.den*other.den
    return f"{temp_num}/{temp_den}"

  def __truediv__(self,other):
    temp_num=self.num*other.den
    temp_den=self.den*other.num
    return f"{temp_num}/{temp_den}"

  def __pow__(self, power):
      temp_num = self.num ** power
      temp_den = self.den ** power
      return Fraction(temp_num, temp_den)
```

```python
f1 = Fraction(1, 2)
f2 = Fraction(3, 4)

print(f1 + f2)    # 10/8
print(f1 - f2)    # -2/8
print(f1 * f2)    # 3/8
print(f1 / f2)    # 4/6
print(f1 ** 2)    # 1/4
```

# **Create new data type with the name of the Line**

```python
class Line:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def __str__(self):
        return f"{self.a}x+{self.b}y+{self.c}"
```

```python
l1=Line(1,1,2)
print(l1)
```

```python
l2=Line(2,4,-16)
print(l2)
```

# **Create new data type with the name of the Point**

```python
class Point:
    def __init__(self,x,y):
        self.x_co=x
        self.y_co=y
    def __str__(self):
        return f"<{self.x_co},{self.y_co}>"

    def __add__(self,other):
        temp_x=self.x_co+other.x_co
        temp_y=self.y_co+other.y_co
        return f"<{temp_x},{temp_y}>"

    def equlidiean_distance(self,other):
        return f"{((((self.x_co-other.x_co)**2)+((self.y_co-other.y_co)**2))**0.5):.2f}"
```

### Write OOP classes to handle the following scenarios:

- A user can create and view 2D coordinates
- A user can find out the distance between 2 coordinates
- A user can find find the distance of a coordinate from origin
- A user can check if a point lies on a given line
- A user can find the distance between a given 2D point and a given line

```python
class Point:
    def __init__(self,x,y):
        self.x_co=x
        self.y_co=y
    def __str__(self):
        return f"<{self.x_co},{self.y_co}>"

    def __add__(self,other):
        temp_x=self.x_co+other.x_co
        temp_y=self.y_co+other.y_co
        return f"<{temp_x},{temp_y}>"

    def equlidiean_distance(self,other):
        # return f"{((((self.x_co-other.x_co)**2)+((self.y_co-other.y_co)**2))**0.5):.2f}"
        return round(((((self.x_co-other.x_co)**2)+((self.y_co-other.y_co)**2))**0.5),2)

class Line:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def __str__(self):
        return f"{self.a}x+{self.b}y+{self.c}=0"

    def point_lie_on_line(self,other):
        # ex_value={self.a}*{other.x_co}+{self.b}*{other.y_co}+{self.c} # {} creates a set, not a number.
        ex_value = self.a * other.x_co + self.b * other.y_co + self.c
        if ex_value==0:
            return "point lies on a line"
        else:
            return  "Point does`t lies on a line"

    def shortest_distance(self,other):
        return round(abs(self.a*other.x_co + self.b*other.y_co + self.c)/(self.a**2 + self.a**2)**0.5,2)
```
