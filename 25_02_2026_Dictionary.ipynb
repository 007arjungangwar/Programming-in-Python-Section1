1. **No Indexing**: No index support.
2. **Mutable**: Can modify post-creation.
3. **Keys**: Immutable (str, num, tuple); **Values**: Any type.
4. **Unique Keys**: Keys must be unique; overwrite on reassignment.

### **Find the character count in the word By using the Dictionary**

````python
word="sliceeliss"
for i in word:
  print(f"word: {i} count: {word.count(i)}")
````

````python
Word="sliceeliss"
dict1={}
for i in Word:
    if i not in dict1:
        dict1[i]=1
    else:
        dict1[i]+=1

dict1
````

````python
Word="sliceeliss"
dict2={}
for i in Word:
    if i in dict2:
        dict2[i]+=1
    else:
        dict2[i]=1

dict2
````

````python
dict2.items()
````

````python
for key,value in dict2.items():
    print(f"char: {key} is coming,{value} times")
````

## **Find the number whose has the maximum digit sum in a list**

````python
list1=[235, 78, 901, 456, 1234]
dict1={}
for i in list1:
    sum=0
    num=i
    original_num=num
    while num>0:
        last_digit=num%10
        num//=10
        sum+=last_digit
    dict1[original_num]=sum
dict1
````

````python
for i in dict1.items():
    print(i)
````

````python
max_value=max([i[1] for i in dict1.items()])
[i for i in dict1.items() if i[1]==max_value]
````

````python
max_value=max([i[1] for i in dict1.items()])
[i[0] for i in dict1.items() if i[1]==max_value]
````

````python
max_value=max(dict1,key=dict1.get)
max_value
````

````python
min_value=min(dict1,key=dict1.get)
min_value
````

## Dictionary Comprehension

```python
{key : value for var in iterable}
```

Creates dicts from iterables.

**Example:**

`squares_dict = {x: x**2 for x in range(5)}`

Maps numbers to squares.

Supports conditions & expressions.

````python
D = {"Name":"Arjun", "Gender":"Male", "Age":28}
D
````

````python
D.items()
````

````python
D1 = {key:value for key, value in D.items() if len(key)>3}
D1
````

````python
L = [1, 2, 3, 4, 5, 6, 7]
````

````python
D2 = {item : item**2 for item in L}
D2
````

````python
D2 = {item : item**2 for item in L if item%2 == 0}
D2
````

## **Print 1st 10 nums & squares**

````python
# Print 1st 10 nums & squares
{i:i**2 for i in range(1, 11)}
````

````python
distances = {'delhi':1000, 'mumbai':2000, 'bangalore':3000}
print(distances.items())
````

````python
# using existing dict
distances = {'delhi':1000, 'mumbai':2000, 'bangalore':3000}
{key:value*0.6 for (key, value) in distances.items()}
````

### **using zip**

````python
# using zip
days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
temp_C = [30.5, 32.6, 31.8, 33.4, 29.8, 30.2, 29.9]
{i:j for (i, j) in zip(days, temp_C)}
````

````python
# using if condition
products = {'phone':10, 'laptop':0, 'charger':32, 'tablet':0}
{key:value for (key, value) in products.items() if value>0}
````

## **Nested Comprehension**

````python
# Nested Comprehension
# Print multiplication tables for 2 to 4
{i:{i*j for j in range(1, 11)} for i in range(2, 5)}
````

````python
# Nested Comprehension
# Print multiplication tables for 2 to 4
{i:{j:i*j for j in range(1, 11)} for i in range(2, 5)}
````

````python
{
    2:{1:2, 2:4, 3:6, 4:8},
    3:{1:3, 2:6, 3:9, 4:12},
    4:{1:4, 2:8, 3:12, 4:16}
}
````
